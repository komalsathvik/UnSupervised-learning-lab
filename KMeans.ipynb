```python
import seaborn as sns
from sklearn.datasets import make_blobs
```

```python
x,y=make_blobs(
    n_samples=1000,
    n_features=2,
    centers=4,
    random_state=42
)
```

```python
sns.scatterplot(x=x[:,0],y=x[:,1] )
```

```python
from sklearn.cluster import KMeans
```

```python
k=4
kmeans=KMeans(
    n_clusters=k,
    random_state=42
)
```

```python
labels=kmeans.fit_predict(x)
labels
```

```python
sns.scatterplot(x=x[:,0],y=x[:,1],c=labels)
```
